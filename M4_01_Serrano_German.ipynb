* Cargar como dataframe de pandas el csv imdb_titulos.csv y mostrar sus 5 primeros registros
* Cargar como dataframe de pandas el csv imdb_elenco.csv y mostrar sus 5 primeros registros
* Mostrar el número de registros del dataframe de titulos
* Mostrar el número de registros del dataframe de elenco
* Mostrar las 5 peliculas más antiguas del listado de titulos
* Mostrar las peliculas que en el titulo tienen la palabra "Dracula". También mostrar el número total de peliculas que coincidan con este requisito
* Mostrar los 10 titulos más comunes (que más se repiten)
* Mostrar cual fue la primer pelicula hecha titulada "Romeo and Juliet" 
* Listar todas las peliculas que contengan la palabra "Exorcist" ordenadas de la más vieja a la más reciente
* Mostrar cuantas peliculas fueron hechas en el año 1950
* Mostrar cuantas peliculas fueron hechas de 1950 a 1959
* Mostrar todos los roles o papeles que hubo en la pelicula "The Godfather". También mostrar el número total de coincidencias
* Mostrar el elenco completo ordenado por la clasificacion "n" de la pelicula "Dracula" de 1958
* Mostrar cuantos papeles de "Bruce Wayne" han sido hechos en la historia de las peliculas
* Mostrar cuantos papeles ha hecho "Robert De Niro" en su carrera
* Listado de papeles como protagonista (n=1) que tuvo el actor "Charlton Heston" en la década de los 60's, ordenado por año de forma descendente
* Mostrar cuantos papeles para actores hubo en la década de los 50's
* Mostrar cuantos papeles para actrices hubo en la década de los 50's

```python
import pandas as pd

pd.set_option('display.max_columns', None)

df_titles = pd.read_csv('imdb_titulos.csv')
df_cast = pd.read_csv('imdb_elenco.csv')

print("Columnas títulos:", df_titles.columns.tolist())
print("Columnas elenco:", df_cast.columns.tolist())

possible_title_cols = ['primaryTitle','title','originalTitle','título','name']
title_col = next((c for c in possible_title_cols if c in df_titles.columns), df_titles.columns[0])

possible_year_cols = ['startYear','year','release_year']
year_col = next((c for c in possible_year_cols if c in df_titles.columns), None)

print('\n5 primeros títulos:')
display(df_titles.head())
print('\n5 primeros registros de elenco:')
display(df_cast.head())

print(f"\nNúmero de registros en titulos: {len(df_titles)}")
print(f"Número de registros en elenco: {len(df_cast)}")

if year_col:
    oldest = df_titles.dropna(subset=[year_col]).sort_values(year_col).head(5)
    print('\n5 películas más antiguas:')
    display(oldest[[title_col, year_col]])
else:
    print('\nNo se encontró columna de año en titulos.')

dracula_mask = df_titles[title_col].str.contains('Dracula', case=False, na=False)
dracula_df = df_titles[dracula_mask]
print(f"\nPelículas con 'Dracula' en el título: {len(dracula_df)}")
display(dracula_df[[title_col] + ([year_col] if year_col else [])].head())

print('\n10 títulos más comunes:')
display(df_titles[title_col].value_counts().head(10))

romeo_df = df_titles[df_titles[title_col] == 'Romeo and Juliet']
if year_col and not romeo_df.empty:
    first_romeo = romeo_df.sort_values(year_col).iloc[0]
    print("\nPrimera película titulada 'Romeo and Juliet':")
    display(first_romeo[[title_col, year_col]])
else:
    print("\nNo se encontró 'Romeo and Juliet' en el dataset o falta columna de año.")

exorcist_df = df_titles[df_titles[title_col].str.contains('Exorcist', case=False, na=False)]
if year_col:
    exorcist_sorted = exorcist_df.dropna(subset=[year_col]).sort_values(year_col)
    print('\nPelículas que contienen "Exorcist" (de más vieja a más reciente):')
    display(exorcist_sorted[[title_col, year_col]])
else:
    display(exorcist_df[[title_col]])

if year_col:
    movies_1950 = df_titles[df_titles[year_col] == 1950]
    movies_50s = df_titles[df_titles[year_col].between(1950, 1959)]
    print(f"\nPelículas en 1950: {len(movies_1950)}")
    print(f"Películas 1950-1959: {len(movies_50s)}")
else:
    print('\nNo se pudo calcular películas por año (columna año no encontrada).')

possible_person_cols = ['name','primaryName','actorName']
possible_title_cast_cols = ['title','primaryTitle','movie_title']
possible_year_cast_cols = ['startYear','year','release_year']
possible_char_cols = ['character','characters','role']

person_col = next((c for c in possible_person_cols if c in df_cast.columns), None)
title_cast_col = next((c for c in possible_title_cast_cols if c in df_cast.columns), None)
year_cast_col = next((c for c in possible_year_cast_cols if c in df_cast.columns), None)
char_col = next((c for c in possible_char_cols if c in df_cast.columns), None)

n_col = 'n' if 'n' in df_cast.columns else None

if title_cast_col:
    godf = df_cast[df_cast[title_cast_col].str.contains('The Godfather', case=False, na=False)]
else:
    godf = df_cast[df_cast.apply(lambda r: r.astype(str).str.contains('The Godfather', case=False).any(), axis=1)]

print(f"\nRoles en 'The Godfather': {len(godf)}")
display(godf.head())

if title_cast_col and year_cast_col and n_col:
    dracula_1958 = df_cast[(df_cast[title_cast_col].str.contains('Dracula', case=False, na=False)) & (df_cast[year_cast_col] == 1958)]
    dracula_1958_sorted = dracula_1958.sort_values(n_col)
    print('\nElenco de "Dracula" (1958) ordenado por n:')
    display(dracula_1958_sorted)
else:
    print('\nNo se pudo ordenar elenco de "Dracula" por n (falta columna).')

if char_col:
    bw = df_cast[df_cast[char_col].str.contains('Bruce Wayne', case=False, na=False)]
    print(f"\nCantidad de papeles de 'Bruce Wayne': {len(bw)}")
else:
    print('\nNo se encontró columna de personajes para buscar "Bruce Wayne".')

if person_col:
    rd = df_cast[df_cast[person_col].str.contains('Robert De Niro', case=False, na=False)]
    print(f"\nCantidad de papeles de Robert De Niro: {len(rd)}")
else:
    print('\nNo se encontró columna de nombre de actor en el dataset de elenco.')

if person_col and n_col and year_cast_col:
    ch = df_cast[(df_cast[person_col].str.contains('Charlton Heston', case=False, na=False)) & (df_cast[n_col] == 1) & (df_cast[year_cast_col].between(1960, 1969))]
    ch_sorted = ch.sort_values(year_cast_col, ascending=False)
    print('\nPapeles como protagonista (n=1) de Charlton Heston en los 60s:')
    display(ch_sorted)
else:
    print('\nNo se pudo calcular papeles protagonistas de Charlton Heston en los 60s (faltan columnas).')

possible_category_cols = ['category','job','roleType']
cat_col = next((c for c in possible_category_cols if c in df_cast.columns), None)

if cat_col and year_cast_col:
    actors_50s = df_cast[(df_cast[cat_col].str.contains('actor', case=False, na=False)) & (df_cast[year_cast_col].between(1950, 1959))]
    actresses_50s = df_cast[(df_cast[cat_col].str.contains('actress', case=False, na=False)) & (df_cast[year_cast_col].between(1950, 1959))]
    print(f"\nPapeles de actores en los 50s: {len(actors_50s)}")
    print(f"Papeles de actrices en los 50s: {len(actresses_50s)}")
else:
    print('\nNo se encontró columna de categoría para contar actores/actrices por década.')
```
